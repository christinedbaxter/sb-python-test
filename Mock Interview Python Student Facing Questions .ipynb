<H2> Challenge Question </H2>

<h4>Q5. Which Job Role has the highest <i> proportion </i> of individuals who earn  more than 50K?</h4>

<b> Put your code below </b>

<h5>5.1 Individuals who earn >50K</h5>

```python
# Number of individuals earning above 50K
n_greater_50k = stripped_df[stripped_df['Earnings'] == '>50K']['Job-Role'].value_counts()

# Number of individuals below or equal to 50K
n_at_most_50k = stripped_df[stripped_df['Earnings'] == '<=50K']['Job-Role'].value_counts()

# Percentage of individuals earning above 50K
greater_percent =  round(n_greater_50k / (n_greater_50k + n_at_most_50k) * 100, 2)

# Print the highest percent value in 'greater_percent' series
print("Highest proportion: {}%".format(greater_percent.max()))

# Print the highest percentage value and the corresponding Job-Role where the percentage is achieved
print("Job-Role with highest proportion of individuals earning above 50K: {}".format(greater_percent.idxmax()))
```

```python
# Plotting the data
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(14,6))

# Create a bar plot for count of individuals earning above 50K
n_greater_50k.plot(kind='barh', ax=ax[0], color='skyblue')
ax[0].set_title('Individuals Earning >50K (n=7,841)')
ax[0].set_xlabel('Number of Individuals')
ax[0].set_ylabel('Job-Role')

# Create a bar plot for percentage of individuals earning above 50K sort descending
greater_percent.sort_values(ascending=True).plot(kind='barh', ax=ax[1], color='salmon')
ax[1].set_title('Percentage of Individuals Earning >50K')
ax[1].set_xlabel('Percentage')
ax[1].set_ylabel('Job-Role')

# Display the plots
plt.tight_layout()
plt.show()
```

<h5>5.2 US High School Graduates who earn >50K</h5>

```python
# Determine job role with highest proportion of US high school graduates earning above 50K
high_income_job_role = (count_gt_50k / (count_lte_50k + count_gt_50k)).sort_values(ascending=False).index[0]

# Percentage of US high school graduates earning above 50K
high_income_percent = round(count_gt_50k / (count_lte_50k + count_gt_50k) * 100, 2)

# Print the highest percentage value and the corresponding Job-Role where the percentage is achieved
print("Highest proportion: {}%".format(high_income_percent.max()))

# Print the Job Role with highest proportion of US high school graduates earning above 50K
print("Job Role with highest proportion of US high school graduates earning above 50K: {}".format(high_income_job_role))
```

```python
# Plotting the data
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(14,6))

# Create a bar plot for count of US high school graduates earning above 50K
count_gt_50k.plot(kind='barh', ax=ax[0], color='skyblue')
ax[0].set_title('US High School Graduates Earning >50K (n=2,006)')
ax[0].set_xlabel('Number of Individuals')
ax[0].set_ylabel('Job-Role')

# Create a bar plot for percentage of US high school graduates earning above 50K sort descending
high_income_percent.sort_values(ascending=True).plot(kind='barh', ax=ax[1], color='salmon')
ax[1].set_title('Percentage of US High School Graduates Earning >50K')
ax[1].set_xlabel('Percentage')
ax[1].set_ylabel('Job-Role')

# Display the plots
plt.tight_layout()
plt.show()
```
