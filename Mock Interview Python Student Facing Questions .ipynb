<b> Q3. The Job Role Columns holds the job information for each individual in this census snapshot. Using this column, create a Bar Chart that shows the count of 'Unique' Jobs per Job Group in the "Job-Role" Column in ascending order, as per the provided image below </b>


<b> Put your code below </b>

```python
# Find the count of unique values in 'Job-Role' column
filtered_adult_census_df['Job-Role'].nunique()
```

```python
# Find the count of unique values in 'Job-Role' column per Job Group
job_counts = filtered_adult_census_df['Job-Role'].value_counts().sort_values()

# Visualize the counts of 'Job-Role' column per Job Group
job_counts.plot(kind='barh', figsize=(10,10))

# Add the title
plt.title('Counts of Job-Role per Job Group')

# Add x-label and y-label
plt.xlabel('Count')
plt.ylabel('Job-Role')

# Display the plot
plt.show()
```
